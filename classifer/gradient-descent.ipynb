```python
import numpy as np 

dataset = np.arange(0 , 100)
dataset
```

```python
dataset = np.expand_dims(dataset , 1)
dataset
dataset = np.repeat(dataset,2,axis = 1 )
dataset[: , 1] = 0
dataset
```

```python

alpha = 0.0001
def calc_gradient(w):
    m = len(dataset)    
    h = w * np.arange(100)
    y = dataset[:,1]
    error = h - y
    gradient =  np.sum(error * np.arange(100)) / m
    return gradient
w = 3
for i in range(100):     
    gradient = calc_gradient(w)
    print("gradient" , gradient)
    w = w - alpha * gradient
    print("w" , w)
print("finally" , w)
```
