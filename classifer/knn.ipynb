# 最小距离分类器 (KNN)

```python
import  tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
(train_images, train_labels) , (test_images , test_labels) = tf.keras.datasets.mnist.load_data()
```

```python
# 让我们来看看 训练集的样本
plt.imshow(train_images[3])
print(train_labels[3])
```

```python
plt.imshow(train_images[30])
print(train_labels[30])
```

```python
plt.imshow(train_images[300])
print(train_labels[300])
```

```python
# 取 60000个训练集
train_images = train_images.reshape(60000 , 28 * 28)
```

```python
# 取10000个测试集
test_images = test_images.reshape(10000,28 * 28)
```

```python
# 归一化
train_images = train_images / 255 
test_images = test_images / 255
```

```python
test = 50 
right = 0 
for i in range(test):
    test_data = test_images[i]
    train_data = train_images[:]
    # L1距离
    distance = np.argmin(np.sum(np.abs(test_data - train_data) , axis = 1))
    predict = np.argmax(train_labels[distance])

    real = np.argmax(test_labels[i])
    if predict == real : 
        i= i+1 

print("准确率" ,i / test)
```
